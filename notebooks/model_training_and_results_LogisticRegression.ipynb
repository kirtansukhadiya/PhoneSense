Importing necessary libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

importing dataset

```python
dataset_X = pd.read_csv('../data/processed/mobile_addiction_cleaned_final.csv')
dataset_y = pd.read_csv('../data/processed/mobile_addiction_target.csv')
y = dataset_y.values.ravel()
```

Dividing into training set and test set

```python
X_train, X_test, y_train, y_test = train_test_split(dataset_X, y, test_size=0.2, random_state=42)
```

Training modal on logistic regression

```python
classifier = LogisticRegression(random_state=0, max_iter= 1000)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

Evaluating accuracy

```python
cm = confusion_matrix(y_test, y_pred)
print(accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", cm)

accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))  
```
